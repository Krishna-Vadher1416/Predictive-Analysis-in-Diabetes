```python
# Insulin -> Right skewed distribution
df['Insulin'] = df['Insulin'].replace(0, df['Insulin'].median())
```

```python
# Pregnancies -> Right skewed distribution
df['Pregnancies'] = df['Pregnancies'].replace(0, df['Pregnancies'].median())
```

```python
df.columns
```

```python
# DiabetesPedigreeFunction -> Right skewed distribution
df['DiabetesPedigreeFunction'] = df['DiabetesPedigreeFunction'].replace(0, df['DiabetesPedigreeFunction'].median())
# Age -> Right skewed distribution
df['Age'] = df['Age'].replace(0, df['Age'].median())
# Glucose -> Normal distribution
df['Glucose'] = df['Glucose'].replace(0, df['Glucose'].mean())
# BloodPressure -> Normal  distribution
df['BloodPressure'] = df['BloodPressure'].replace(0, df['BloodPressure'].mean())
# SkinThickness -> Right skewed  distribution
df['SkinThickness'] = df['SkinThickness'].replace(0, df['SkinThickness'].median())
# BMI -> Normal  distribution
df['BMI'] = df['BMI'].replace(0, df['BMI'].mean())
```

```python
df.head(20)
```

```python
## X -> imput features , y-> value
X = df.drop(columns="Outcome", axis = 1)
Y = df["Outcome"]
```

## Outlire Detection

```python
fig, ax = plt.subplots(figsize= (15,15))
sns.boxplot(data = X, ax = ax)
plt.savefig("boxPlot.jpg")
```

```python
X.shape
```

```python
Y.shape
```

```python
cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
for col in cols:
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    mask = (X[col] >= lower_bound) & (X[col] <= upper_bound)
```

```python
X_outlier_detection = X[mask]
Y_outlier_detection = Y[mask]
```

```python
X_outlier_detection.shape
```

```python
Y_outlier_detection.shape
```

## Approch 2 of removing outliers: Quantiles 

```python
X_scaled.reset_index(drop=True, inplace=True)
Y_outlier_detection.reset_index(drop=True, inplace=True)
```

```python
q = X_scaled['Insulin'].quantile(.95)
mask = X_scaled['Insulin'] < q
dataNew = X_scaled[mask]
Y_outlier_detection = Y_outlier_detection[mask]
```

## Model Training 

### Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataNew, Y_outlier_detection, test_size=0.33, random_state=42)
```

### Data Imbalance


1.Oversampling : Minority class and increase that nummber to the majority class



2.Undersampling : Majority class and decrease that number to the minority class 

3.SMOTE : Synthetic data increase the number of samples to the majority class

```python
print(y_train.value_counts())
```

## SMOTE Technique

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Check resampled class distribution
print("\nResampled class distribution:")
print(pd.Series(y_train_resampled).value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
# Initialize Logistic Regression classifier
classification = LogisticRegression()
# Fit the model to the training data
classification .fit(X_train_resampled, y_train_resampled)
```

## Model Prediction

```python
y_predictions = classification.predict(X_test)
print(y_predictions)
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predictions)
```

```python
from sklearn.metrics import classification_report
```

### Healthcare: Recall is very important metric

```python
target_names = ["Non-Diabetic","Diabetic"]
print(classification_report(y_test, y_predictions, target_names= target_names))
```
